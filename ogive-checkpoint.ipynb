```python
import numpy as np
import matplotlib.pyplot as plt

data = np.random.randint(0, 100, 500)
population=[]
for i in range(500):
    num=np.random.randint(1000)
    population.append(num)
print("Population:",population)
sample=list(np.random.choice(population,20))
print("Sample:",sample)

bins = np.arange(0, 110, 10)
freq, bins = np.histogram(sample, bins=bins)
cumulative_freq = np.cumsum(freq)
res = np.flipud(cumulative_freq)
plt.plot(bins[1:], cumulative_freq, color='blue', marker='o', linestyle='-')
plt.xlabel('Data')
plt.ylabel('Cumulative Frequency')
plt.title('More than Ogive Graph of Random Data')
plt.show()
plt.plot(bins[1:], res, color='red', marker='o', linestyle='-')
plt.xlabel('Data')
plt.ylabel('Cumulative Frequency')
plt.title('Less than Ogive Graph of Random Data')
plt.show()
```
