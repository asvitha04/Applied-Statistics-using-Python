```python
import pandas as pd
import numpy as np
import math
import matplotlib.pyplot as plt
%matplotlib inline
dataset=[]
n=int(input("Enter the no. of entries:"))
print("Enter data:")
for i in range(n):
    data=int(input())
    dataset.append(data)
frequency = {}

for item in dataset:
    if item in frequency:
        frequency[item] += 1
    else:
        frequency[item] = 1
print(frequency)
val=list(frequency.keys())
ks=len(dataset)
x_axistitle=input("Enter x axis title:")
plt.hist(dataset,bins=range(2*ks), color='yellow', edgecolor='black', linewidth=1.5)
plt.xlabel(x_axistitle)
plt.ylabel("Frequency")
plt.title("Histogram for frequency of data")
plt.legend(["freq"],loc='best')
plt.grid()
plt.show()
```
