```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from matplotlib import style

titles=[]
values=[]
n=int(input("Enter the number of entries:"))
print("Enter the titles:")
for i in range(n):
    title=input()
    titles.append(title)
print("Enter the values:")
for j in range(n):
    value=int(input())
    values.append(value)
color=[]
print("Enter colors to use in the graph:")
for i in range(n):
    col=input()
    color.append(col)
a=[titles,values]
y_pos = np.arange(len(a[0]))
x_axistitle=input("Enter x axis title:")
y_axistitle=input("Enter y axis title:")
plt.bar(a[0],a[1],color=color,width=0.2, align='center', alpha=1)
plt.xlabel(x_axistitle)
plt.ylabel(y_axistitle)
plt.xticks(y_pos, a[0])
plt.legend(color,loc ="lower right")
```

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from matplotlib import style
titles=[]
values=[]
n=int(input("Enter the number of entries:"))
print("Enter the titles:")
for i in range(n):
    title=input()
    titles.append(title)
print("Enter the values:")
for j in range(n):
    value=int(input())
    values.append(value)
color=[]
print("Enter colors to use in the graph:")
for i in range(n):
    col=input()
    color.append(col)
a=[titles,values]
y_pos = np.arange(len(a[0]))
x_axistitle=input("Enter x axis title:")
y_axistitle=input("Enter y axis title:")
plt.bar(a[0],a[1],color=color)
plt.xlabel(x_axistitle)
plt.ylabel(y_axistitle)
plt.xticks(y_pos, a[0])
plt.show()

plt.pie(values,labels=titles,startangle=90,colors=color,autopct="%2.1f%%")
plt.legend(title=x_axistitle)
plt.show()
```
